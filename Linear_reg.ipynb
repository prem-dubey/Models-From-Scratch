# Linear Regression 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python

x = np.array([1,2,3,4,5])
y = np.array([3,5,7,9,11])
plt.plot(x,y)
plt.title('expected line')
plt.show()
```

```python
# asumming a linear model for this traiing y = wx + b
# initializing the values of w and b
w = 0
b = 0 
alpha = 0.1
```

```python
# implementing gradient descent 
def gradDesJw(m  , alpha , w , b, x , y):
    Jw = 0
    for i in range(m):
        Jw += (w*x[i] + b - y[i])*x[i]
       
    return Jw / m

def gradDesJb(m , alpha , w  , b , x  , y ):
    Jb = 0
    for i in range(m):
        Jb += (w*x[i] + b - y[i])

    return Jb / m


```

```python
w_des = alpha * gradDesJw(5,alpha,w,b,x,y) 
b_des = alpha * gradDesJb(5,alpha,w,b,x,y)
print(w_des , b_des)

t = 500;
while(t >= 0):
    w = w - w_des
    b = b - b_des
    w_des = alpha * gradDesJw(5,alpha,w,b,x,y) 
    b_des = alpha * gradDesJb(5,alpha,w,b,x,y) 
    print(w_des, b_des)
    t = t-1

print(w,b)



```

```python
y_pred = w*x + b # calculating all of the predicted values 

plt.plot(x,y,color='red')
plt.plot(x,y_pred,color='blue')
plt.suptitle('Comparing the Lines')
plt.show()

# plotting the two lines and comparing 
plt.subplot(1,2,1)
plt.plot(x,y)

# comparing the two lines 
plt.subplot(1,2,2)
plt.plot(x,y_pred)

plt.suptitle('comparing the two plots')

plt.show()
```
